```python
import numpy as np
import matplotlib.pyplot as plt
import sempler
from sempler import functions, noise
```

```python
A = np.array([[0, 1], [0 ,0]])
noise_distributions = [noise.uniform(-3,3), noise.normal(0, 0.01)]
functions = [sempler.functions.null, np.exp]
anm = sempler.ANM(A, functions, noise_distributions)
```

```python
X = anm.sample(100)
plt.scatter(X[:,0], X[:,1])
```

```python
X = anm.sample(100, shift_interventions = {1: noise.normal()})
plt.scatter(X[:,0], X[:,1])
```

```python
X = anm.sample(100, do_interventions = {0: noise.normal()})
plt.scatter(X[:,0], X[:,1])
```

```python
X = anm.sample(100, do_interventions = {1: noise.normal()})
plt.scatter(X[:,0], X[:,1])
```

```python
X = anm.sample(100, do_interventions = {0: noise.normal(), 1: noise.normal()})
plt.scatter(X[:,0], X[:,1])
```
